# Grove Buzzer v1.2

This example shows how to use the [Grove Buzzer v1.2](http://www.seeedstudio.com/wiki/Grove_-_Buzzer).

A Grover Buzzer, and Pmod Stickit connector are required. 

The Grove Buzzer will emit a tone when the output is high. A frequency tone can be generated by varying the time the output is held high. 

The Pmod Stickit is connected to IOP1 (JB) and the Grove Buzzer is connected to GR 1 of the Stickit. 

```python
from pynq import Overlay

Overlay("pmod.bit").download()
```

### 1. Illustrate playing a pre-defined melody

```python
from pynq.pmods import Grove_Buzzer

# Stickit is connected to PMOD 1 labeled JB
# Grove Buzzer is connected to GR 1 
grove_buzzer = Grove_Buzzer(1, 1)
grove_buzzer.play_melody()
```

### 2. Generate a tone of desired period and for a desired number of times
The tone_period is in microseconds and the 50% duty cycle will be generated for the given tone_period 

```python
from pynq.pmods import Grove_Buzzer

tone_period = 1200
num_cycles = 500
grove_buzzer = Grove_Buzzer(1, 1)
grove_buzzer.play_tone(tone_period,num_cycles)
```
