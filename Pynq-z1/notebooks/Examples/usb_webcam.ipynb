# USB Webcam

In this XPP Notebook, a simple USB web camera has been attached to the Zybo board.  A picture is taken using fswebcam and that image can then be manipulated using the Python Image Library (Pillow).

This flow is demonstrated using a Logitech USB HD Webcam C270 attached to Zybo.  The camera can be bought from the 3rd link below.

#### References
http://pillow.readthedocs.org/en/3.1.x/handbook/tutorial.html<br>
http://manpages.ubuntu.com/manpages/lucid/man1/fswebcam.1.html <br>
http://www.logitech.com/en-us/product/hd-webcam-c270

```python
from PIL import Image as PIL_Image

orig_img_path = '/home/xpp/jupyter_notebooks/Examples/data/webcam.jpg'
!fswebcam  --no-banner --save {orig_img_path} -d /dev/video0 2> /dev/null

img = PIL_Image.open(orig_img_path)
img
```

```python
bw_img = img.convert("L")
bw_img
```

```python
rot_img = img.rotate(45)
rot_img
```
