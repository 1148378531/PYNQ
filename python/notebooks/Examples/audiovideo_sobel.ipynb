## Demo : Sobel Filter

This demo will take the HDMI input and convert it into an image. 
This image wil be transformed into a gray-scale image, and later processed by the Sobel filter.

Since the image is processed by the PS, it may take a long time to get the results.

```python
from time import sleep
from pynq import Overlay
from pynq.video import Frame, HDMI
from IPython.display import Image

ol = Overlay("audiovideo.bit")
ol.download()

hdmi=HDMI('in')
sleep(20) 
hdmi.start()

frame = hdmi.frame()
height = hdmi.frame_height()
width = hdmi.frame_width()

orig_img_path = '/home/xpp/jupyter_notebooks/data/orig.jpg'
frame.save_as_jpeg(orig_img_path)
Image(filename=orig_img_path)
```

```python
for y in range(0, height):
    for x in range(0, width):
        gray = round((0.299*frame[x,y][0]) + (0.587*frame[x,y][1]) 
                      + (0.114*frame[x,y][2]))
        frame[x,y] = (gray, gray, gray)

gray_img_path = '/home/xpp/jupyter_notebooks/data/gray.jpg'
frame.save_as_jpeg(gray_img_path)
Image(filename=gray_img_path)
```

```python
#Compute the Sobel Filter output
#Gx = -1 0 +1     Gy= +1 +2 +1
#     -2 0 +2          0  0  0
#     -1 0 +1         -1 -2 -1

sobel = Frame(width, height)

for y in range(1,height-1):
    for x in range(1,width-1):
        gx = abs(-frame[x-1,y+1][0] + frame[x+1,y+1][0] -2*frame[x-1,y][0] +2*frame[x+1,y][0] 
                 -frame[x-1,y-1][0] + frame[x+1,y-1][0])
        gy = abs(frame[x-1,y+1][0] + 2*frame[x,y+1][0] + frame[x+1,y+1][0] -frame[x-1,y-1][0]
                 -2*frame[x,y-1][0] - frame[x+1,y-1][0])
        grad = gx + gy
        if grad > 255:
            grad = 255                    
        sobel[x,y] = (grad,grad,grad)
        
sobel_img_path = '/home/xpp/jupyter_notebooks/data/sobel.jpg'
sobel.save_as_jpeg(sobel_img_path)

del frame
del sobel
del hdmi

ol = Overlay("pmod.bit")
ol.download()

Image(filename=sobel_img_path)
```
